```python
# state variable mapping
mydict={1:'AL', 2:'AK', 3:'AZ', 4:'AR', 5:'CA', 6:'CO', 7:'CT', 8:'DE', 9:'DC', 10:'FL', 11:'GA', 12:'HI', 13:'ID',
14:'IL', 15:'IN', 16:'IA', 17:'KS', 18:'KY', 19:'LA', 20:'ME', 21:'MD', 22:'MA', 23:'MI', 24:'MN', 25:'MS', 26:'MO', 27:'MT',
28:'NE', 29:'NV', 30:'NH', 31:'NJ', 32:'NM', 33:'NY', 34:'NC', 35:'ND', 36:'OH', 37:'OK', 38:'OR', 39:'PA', 40:'RI', 41:'SC',
42:'SD', 43:'TN', 44:'TX', 45:'UT', 46:'VT', 47:'VA', 48:'WA', 49:'WV', 50:'WI', 51:'WY'}
```

```python
type(MMI2019Q2_benchmark['State-Headquarters'][0])==float
```

```python
def renameState(x):
    if x in mydict.keys():
        return mydict.get(x)
    else:
        return x
```

```python
# rename the state column
MMI2019Q2_benchmark['State-Headquarters']=MMI2019Q2_benchmark['State-Headquarters'].apply(renameState)
```

```python
def recodeIndustry(x):
    if x==1:
        return 1
    elif x==16:
        return 2
    elif x==3:
        return 3
    elif (x==9) | (x==10) | (x==11):
        return 4
    elif x==14:
        return 5
    elif (x==6) | (x==15):
        return 6
    elif x==12:
        return 7
    elif (x==4) | (x==5):
        return 8
    elif (x==2) | (x==7) | (x==8)| (x==13):
        return 9
    elif x==17:
        return ''
```

```python
MMI2019Q2_benchmark['Industry']=MMI2019Q2_benchmark[MMI2019Q2_benchmark['Wave']==30]['Industry'].apply(recodeIndustry)
```
