```python
outliers=[]
def find_outliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)  
    for x in data:
        z_score=(x-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(x)
    outlierSet=set(outliers)
    return outlierSet

def find_outliers_list(myDF):
    outlierList=[]
    outliers=[]
    dic={}
    threshold=3
    columns=list(myDF)
    for column in columns:
        mean=np.mean(myDF[column])
        std=np.std(myDF[column])  
        for x in myDF[column]:
            z_score=(x-mean)/std
            if np.abs(z_score)>threshold:
                outliers.append(x)
        outlierSet=set(outliers)
        outlierList.append(sorted(outlierSet))
    return outlierList

q2ar1Outliers=find_outliers(newGuest.s2ar1)
q2ar2Outliers=find_outliers(newGuest.s2ar2)
s2eOutliers=find_outliers(newGuest.s2e)
G12bOut=find_outliers(newGuest.G12b)
G12cOut=find_outliers(newGuest.G12c)

outList=find_outliers_list(questions)


print('q2ar1Outliers:',sorted(q2ar1Outliers))
print('q2ar2Outliers:',sorted(q2ar2Outliers))
print('s2eOutliers:',sorted(s2eOutliers))
print('G12bOut:',sorted(G12bOut))
print('G12cOut:',sorted(G12cOut))
print(outList)
```
